# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width  and height

# plot data
ax.plot(x,y)

# customise plot
ax.set(
  title="Simple plot",
  xlabel='x-axis',
  ylabel='y-axis'
       )
 # 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

```python
# Make a plot from dictionary

nut_butter_prices = {'Almound butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title='Dans Nut Butter Store',
       ylabel='Price($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# make some data for histogram and plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
#subplot option 1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# subploats option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
#plot to each different index
ax[0,0].plot(x,x/3);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
# Make a dataframe
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2024', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]','',regex=True)
car_sales
```

```python
# Remove the last two zeros
car_sales['Price'].astype(str)
car_sales['Price'] = car_sales['Price'].str[:-2] 
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2020' ,periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
# Lets plot the  total sales
car_sales.plot(x='Sale Date' , y='Total Sales');
```

```python
car_sales['Price'] = car_sales['Price'].astype(int)
car_sales.plot(x='Odometer (KM)', y='Price' ,kind='scatter');
```
